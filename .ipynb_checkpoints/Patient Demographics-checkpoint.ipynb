```python
df['ethnicity'] = df['ethnicity'].replace({'Black/African American': 'African American',
                                           '': 'Other/Unknown/Missing',
                                           'Native American':'Indigenous/Native',
                                           'Other/Unknown':'Other/Unknown/Missing'})
df['age_num'] = df['age_num'].astype(float)
df['ApacheScore'] = df['ApacheScore'].astype(float)
df['daysonvent'] = df['daysonvent'].astype(float)


df['gender'] = df['gender'].replace({ '': 'Unknown'})
df = df.fillna({'ethnicity':'Other/Unknown/Missing', 'gender':'Unknown'})
df = df.rename(columns = {'has_sepsis':'Sepsis', 'HOSP_NUM':'Number_of_Admissions', 'hospitaldischargeyear':'Discharge_Year',
                          'age_num':'Age', 'hosp_mortality':'hospital_mortality'})

df['set'] = df['Discharge_Year'].apply(
    lambda x: '2014-2015' if x in [2014, 2015] else(
              '2020-2021' if x in [2020, 2021] else np.nan
    )
)
```

```python
EquiFlow._clean_missing = lambda self: None

eqfl = EquiFlow(
    data=df,
    initial_cohort_label='Admitted from 2014-2021',
    categorical=['gender', 'ethnicity', 'Sepsis', 'hospital_mortality']  
)

eqfl.add_exclusion(
    mask=(df.Number_of_Admissions == 1),
    exclusion_reason='Not First Admission',
    new_cohort_label=''
)

eqfl.add_exclusion(
    mask=(df.Sepsis == 1),
    exclusion_reason='Non-Septic Patients',
    new_cohort_label=''
)


eqfl.add_exclusion(
    mask=(df.ApacheScore.notna()),
    exclusion_reason='Missing APACHE Score',
    new_cohort_label=''
)


eqfl.add_exclusion(
    mask=(df.hospital_mortality.notna()),
    exclusion_reason='Missing Mortality Data',
    new_cohort_label=''
)

eqfl.add_exclusion(
    mask=(df.predicted_hospital_mortality >= 0),
    exclusion_reason='Predicted Hopsital\nMortality =-1',
    new_cohort_label=''
)





eqfl.plot_flows()
```

```python
year_groups = {
    "2014–2015": [2014, 2015],
    "2020–2021": [2020, 2021]
}

results = {}

for ethnicity in df['ethnicity'].dropna().unique():
    results[ethnicity] = {}
    
    for version in ['IV', 'IVa', 'IVb']:
        for group_name, years in year_groups.items():
            subset = df[
                (df['ethnicity'] == ethnicity) &
                (df['Apache_Version'] == version) &
                (df['Discharge_Year'].isin(years))
            ]

            y_true = subset['mortality']
            y_score = subset['predicted_hospital_mortality']

            if len(y_true.unique()) < 2:
                auroc = None  
            else:
                auroc = round(roc_auc_score(y_true, y_score), 2)

            col_name = f"{version} {group_name}"
            results[ethnicity][col_name] = auroc

auroc_table = pd.DataFrame.from_dict(results, orient='index')
auroc_table.index.name = 'Ethnicity'
auroc_table.reset_index(inplace=True)

auroc_table
```

```python

def compute_auroc_ci(data, label, group):
    if data['mortality'].nunique() < 2:
        return None
    auroc = roc_auc_score(data['mortality'], data['predicted_hospital_mortality'])
    boot_scores = []
    for _ in range(1000):
        boot = resample(data)
        if boot['mortality'].nunique() == 2:
            boot_scores.append(roc_auc_score(boot['mortality'], boot['predicted_hospital_mortality']))
    return {
        'Label': label,
        'Group': group,
        'AUROC': auroc,
        'Lower_CI': np.percentile(boot_scores, 2.5),
        'Upper_CI': np.percentile(boot_scores, 97.5)
    }

for apache in df['Apache_Version'].unique():
    df_apache = df[df['Apache_Version'] == apache]
    results = []

    results.append(compute_auroc_ci(df_apache, 'Total', 'total'))

    for s in df_apache['set'].unique():
        results.append(compute_auroc_ci(df_apache[df_apache['set'] == s], s, 'year'))

    for s in df_apache['set'].unique():
        df_set = df_apache[df_apache['set'] == s]
        for eth in df_set['ethnicity'].dropna().unique():
            label = f"{eth} ({s})"
            group = f"ethnicity_{s}"
            res = compute_auroc_ci(df_set[df_set['ethnicity'] == eth], label, group)
            if res:
                results.append(res)

    res_df = pd.DataFrame(results)

    color_map = {
        'total': '#f4a261',
        'year': '#888888',
        'ethnicity_2014-2015': '#2a9d8f',
        'ethnicity_2020-2021': '#9b5de5'
    }
    res_df['color'] = res_df['Group'].map(color_map)

    fig, ax = plt.subplots(figsize=(14, 6))
    x = np.arange(len(res_df)) * 1.5
    y = res_df['AUROC']
    yerr = [y - res_df['Lower_CI'], res_df['Upper_CI'] - y]

    ax.bar(x, y, yerr=yerr, capsize=4, color=res_df['color'], edgecolor='black')
    ax.axhline(res_df.loc[0]['AUROC'], color='black', linestyle='--', linewidth=1, label='Mean AUROC')

    ax.set_xticks(x)
    ax.set_xticklabels(res_df['Label'], rotation=45, ha='right', fontsize=9)
    ax.set_ylabel('AUROC')
    ax.set_title(f'Discrimination by Ethnicity and Year - {apache}')
    ax.set_ylim(0.5, 1.0)
    ax.legend()
    plt.tight_layout()
    plt.show()
```

```python
all_results = []

for apache in df['Apache_Version'].unique():
    df_apache = df[df['Apache_Version'] == apache]
    results = []

    res_total = compute_auroc_ci(df_apache, 'Total', 'total')
    if res_total:
        res_total['Apache_Version'] = apache
        res_total['set'] = 'total'
        results.append(res_total)

    for s in df_apache['set'].unique():
        res_year = compute_auroc_ci(df_apache[df_apache['set'] == s], s, 'year')
        if res_year:
            res_year['Apache_Version'] = apache
            res_year['set'] = s
            results.append(res_year)

    for s in df_apache['set'].unique():
        df_set = df_apache[df_apache['set'] == s]
        for eth in df_set['ethnicity'].dropna().unique():
            label = eth  
            group = f"ethnicity_{s}"
            res_eth = compute_auroc_ci(df_set[df_set['ethnicity'] == eth], label, group)
            if res_eth:
                res_eth['Apache_Version'] = apache
                res_eth['set'] = s
                res_eth['ethnicity'] = eth 
                results.append(res_eth)

    all_results.extend(results)

summary_df = pd.DataFrame(all_results)
```

```python
pivot_df = summary_df.pivot_table(
    index='ethnicity',
    columns=['Apache_Version', 'set'],
    values='AUROC'
)
```
